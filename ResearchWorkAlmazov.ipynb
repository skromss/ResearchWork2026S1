```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import roc_auc_score, confusion_matrix, precision_score, recall_score, RocCurveDisplay
import lightgbm as lgb
import spacy
import re
from nltk.corpus import stopwords
import nltk
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nltk.download('stopwords')
stopwords_ru = set(stopwords.words('russian')) - {'не', 'без', 'ни'}
nlp = spacy.load('ru_core_news_sm')
```

```python
# Функция очистки текста для "Описание диагноза"
def clean_text(text):
    """Очищает текст: приводит к нижнему регистру, удаляет лишние символы и стоп-слова."""
    text = str(text).lower()  # Приведение к строке на случай NaN
    text = text.replace('_', ' ')
    # Удаляем лишние символы, сохраняем русские буквы, латинские буквы и цифры
    text = re.sub(r'[^а-яёa-z0-9 ]', ' ', text)
    doc = nlp(text)
    # Лемматизация и удаление стоп-слов
    lemmas = [token.lemma_ for token in doc if token.text not in stopwords_ru]
    return re.sub(r'\s+', ' ', " ".join(lemmas)).strip()

# Функция извлечения препаратов
def extract_medications(text):
    """Извлекает названия препаратов из текста в нижнем регистре и возвращает их список."""
    if pd.isna(text):
        return []
    # Приведение текста к нижнему регистру
    text = str(text).lower()
    # Регулярное выражение для поиска названий препаратов (слова от 3 букв, могут содержать дефис)
    pattern = r'(?:[а-яёa-z][а-яёa-z-]{2,}(?:\s+[а-яёa-z][а-яёa-z-]{2,})*)'
    matches = re.findall(pattern, text)
    # Расширенный список исключений
    exclude_words = {
        'медикаментозная', 'терапия', 'таб', 'сут', 'вечером', 'утром', 'днем', 'ночь', 'контроль',
        'продолжить', 'явка', 'результатами', 'протокол', 'проба', 'нагрузка', 'достигнутая', 'минуте',
        'минуты', 'минут', 'раз', 'день', 'нед', 'мес', 'мг', 'тб', 'кап', 'форте', 'актив', 'плюс',
        'ад', 'qt', 'mets', 'bruce', 'span', 'style', 'font-size', 'font-style', 'font-weight',
        'text-decoration', 'underline', 'inherit', 'italic', 'color', 'pounceau', 'quinolin', 'yellow',
        'chrono', 'cito', 'ds', 'rp', 'rw', 'pq', 'qtc', 'nt-probnp', 'ntprobnp', 'mb', 'me', 'nf',
        'vasc', 'vit', 'www', 'xl', 'xл', 'аб', 'ав', 'аг', 'аза', 'аат', 'адекватная', 'алат',
        'аллергия', 'аллергическая', 'реакция', 'аллерголога', 'иммунолога', 'обследование',
        'противопоказаны', 'противопоказаний', 'достоверных', 'ишемических', 'изменений', 'ангинозный',
        'приступ', 'дозу', 'увеличить', 'уменьшить', 'добавить', 'отмена', 'отменить', 'наблюдение',
        'кардиолога', 'гастроэнтеролог', 'консультация', 'дообследование', 'цель', 'черед', 'почек',
        'липиды', 'остеоартроз', 'сохранении', 'выше', 'ниже', 'менее', 'более', 'однократном',
        'повышении', 'планово', 'крови', 'препараты', 'железа', 'аализ', 'аденома', 'надпочечника',
        'проведение'
    }
    # Словарь синонимов для препаратов
    synonyms = {
        'валсартан': ['вальсакор', 'диован', 'валз'],
        'лозартан': ['лозап', 'лориста'],
        'бисопролол': ['конкор', 'небилет', 'небиволол', 'небивололум'],
        'метопролол': ['эгилок', 'беталок зок'],
        'аторвастатин': ['ааторвастатин', 'аиорвастатин', 'липримар', 'акорта', 'аторвастатинум'],
        'розувастатин': ['роксера', 'мертенил', 'розулип', 'крестор'],
        'аспирин': ['кардиомагнил', 'аспирин-кардио', 'ааспирин кардио', 'ацетилсалициловая к-та', 'ацетилсалициловая кислота', 'тромбо асс'],
        'периндоприл': ['престариум', 'ко-перинева', 'нолипрел', 'нолипрел-форте'],
        'амиодарон': ['аллапинин', 'аллапенин', 'аллапини', 'аллопинин', 'алларитмин'],
        'омепразол': ['омепrazole', 'нолпаза'],
        'гидрохлортиазид': ['акрипамид', 'акрипамид ретард'],
        'азилсартан': ['азилсаратан', 'азилсартана', 'азилсартан медоксомил', 'азилсартиан', 'эдрби', 'эдарби', 'эдарби кло', 'edarbyclor']
    }
    # Инвертируем словарь синонимов для маппинга
    synonym_map = {}
    for main_name, syn_list in synonyms.items():
        for syn in syn_list:
            synonym_map[syn] = main_name
    synonym_map.update({k: k for k in synonyms.keys()})  # Добавляем основные названия

    # Список известных препаратов для фильтрации
    known_drugs = {
        'валз', 'конкор', 'верошпирон', 'амлодипин', 'физиотенз', 'локрен', 'престариум', 'торасемид',
        'диован', 'лозап', 'раеном', 'атенолол', 'верапамил', 'небилет', 'лориста', 'бисопролол',
        'валсартан', 'ко-перинева', 'периндоприл', 'индапамид', 'вальсакор', 'нифедипин', 'нифекард',
        'теветен', 'эгилок', 'метопролол', 'эксфорж', 'эгипресс', 'рамиприл', 'микардис', 'аторвастатин',
        'розувастатин', 'аспирин', 'клопидогрел', 'ривароксабан', 'ксарелто', 'омепразол', 'аллапинин',
        'адаптол', 'глицин', 'мексидол', 'мексикор', 'циннаризин', 'аспаркам', 'панангин', 'доппельгерц',
        'аквадетрим', 'аллопуринол', 'акридипин', 'акролипид', 'аксамон', 'актовегин', 'аккузид',
        'аккупро', 'аза', 'аденорик', 'аденурик', 'адеметионин', 'аевит', 'азилсартан', 'акатинол',
        'актрапид', 'алирокумаб', 'гидроксизин', 'доксазосин', 'тразодон', 'блопресс', 's-амлодипин',
        'solgar', 'cyanocobalamin', 'депаксин'
    }

    # Фильтрация и маппинг синонимов
    medications = []
    for med in matches:
        if med not in exclude_words and len(med) >= 3:  # Исключаем короткие слова
            # Маппинг на основное название, если это синоним
            main_name = synonym_map.get(med, med)
            if main_name in known_drugs and main_name not in medications:
                medications.append(main_name)
    return medications

# Подсчет антигипертензивных препаратов
def count_antihypertensive_medications(medications):
    """Подсчитывает количество антигипертензивных препаратов в списке."""
    antihypertensive_drugs = {
        'валз', 'конкор', 'верошпирон', 'амлодипин', 'физиотенз', 'локрен', 'престариум', 'торасемид',
        'диован', 'лозап', 'раеном', 'атенолол', 'верапамил', 'небилет', 'лориста', 'бисопролол',
        'валсартан', 'ко-перинева', 'периндоприл', 'индапамид', 'вальсакор', 'нифедипин', 'нифекард',
        'теветен', 'эгилок', 'метопролол', 'эксфорж', 'эгипресс', 'рамиприл', 'микардис', 'азилсартан',
        's-амлодипин', 'блопресс'
    }
    return sum(1 for med in medications if med in antihypertensive_drugs)

# Обработка столбца "Код диагноза" (Label Encoding)
def encode_diagnosis_code(df):
    df['Код диагноза'], _ = df['Код диагноза'].factorize()
    return df

# Предобработка текстового столбца "Описание диагноза"
def preprocess_description(df):
    df['Описание диагноза'] = df['Описание диагноза'].apply(clean_text)
    return df

# Векторизация препаратов
def apply_medication_tfidf(X_train, X_test, X_oot):
    """Векторизует списки препаратов с помощью TF-IDF."""
    # Преобразуем списки препаратов в строки
    X_train['medications_str'] = X_train['medications'].apply(lambda x: ' '.join(x))
    X_test['medications_str'] = X_test['medications'].apply(lambda x: ' '.join(x))
    X_oot['medications_str'] = X_oot['medications'].apply(lambda x: ' '.join(x))
    
    # Применяем TF-IDF
    tfidf = TfidfVectorizer(max_features=50)  # Ограничиваем до 50 признаков
    tfidf_train = tfidf.fit_transform(X_train['medications_str']).toarray()
    tfidf_test = tfidf.transform(X_test['medications_str']).toarray()
    tfidf_oot = tfidf.transform(X_oot['medications_str']).toarray()
    
    # Создаем новые DataFrame с TF-IDF признаками
    tfidf_columns = [f'med_tfidf_{i}' for i in range(tfidf_train.shape[1])]
    X_train_tfidf = pd.DataFrame(tfidf_train, columns=tfidf_columns, index=X_train.index)
    X_test_tfidf = pd.DataFrame(tfidf_test, columns=tfidf_columns, index=X_test.index)
    X_oot_tfidf = pd.DataFrame(tfidf_oot, columns=tfidf_columns, index=X_oot.index)
    
    # Удаляем временный столбец medications_str и medications
    X_train = X_train.drop(columns=['medications_str', 'medications']).join(X_train_tfidf)
    X_test = X_test.drop(columns=['medications_str', 'medications']).join(X_test_tfidf)
    X_oot = X_oot.drop(columns=['medications_str', 'medications']).join(X_oot_tfidf)
    
    return X_train, X_test, X_oot

# TF-IDF векторизация для "Описание диагноза"
def apply_tfidf(X_train, X_test, X_oot):
    tfidf = TfidfVectorizer(max_features=100)  # Ограничим до 100 признаков
    tfidf_train = tfidf.fit_transform(X_train['Описание диагноза']).toarray()
    tfidf_test = tfidf.transform(X_test['Описание диагноза']).toarray()
    tfidf_oot = tfidf.transform(X_oot['Описание диагноза']).toarray()
    
    # Создаем новые DataFrame с TF-IDF признаки
    tfidf_columns = [f'tfidf_{i}' for i in range(tfidf_train.shape[1])]
    X_train_tfidf = pd.DataFrame(tfidf_train, columns=tfidf_columns, index=X_train.index)
    X_test_tfidf = pd.DataFrame(tfidf_test, columns=tfidf_columns, index=X_test.index)
    X_oot_tfidf = pd.DataFrame(tfidf_oot, columns=tfidf_columns, index=X_oot.index)
    
    # Удаляем оригинальный столбец "Описание диагноза" и добавляем TF-IDF признаки
    X_train = X_train.drop(columns=['Описание диагноза']).join(X_train_tfidf)
    X_test = X_test.drop(columns=['Описание диагноза']).join(X_test_tfidf)
    X_oot = X_oot.drop(columns=['Описание диагноза']).join(X_oot_tfidf)
    
    return X_train, X_test, X_oot

# Полная предобработка данных
def preprocess_data(X_train, X_test, X_oot):
    X_train = encode_diagnosis_code(X_train)
    X_test = encode_diagnosis_code(X_test)
    X_oot = encode_diagnosis_code(X_oot)
    
    X_train = preprocess_description(X_train)
    X_test = preprocess_description(X_test)
    X_oot = preprocess_description(X_oot)
    
    X_train, X_test, X_oot = apply_tfidf(X_train, X_test, X_oot)
    X_train, X_test, X_oot = apply_medication_tfidf(X_train, X_test, X_oot)
    
    return X_train, X_test, X_oot
```

```python
# Диагностика данных
print("\nДиагностика данных для PDP:")
for feature in key_features:
    print(f"{feature}:")
    print(f"  Пропуски: {X_test[feature].isna().sum()}")
    print(f"  Уникальные значения: {X_test[feature].nunique()}")
    print(f"  Диапазон: [{X_test[feature].min():.2f}, {X_test[feature].max():.2f}]")

fig, ax = plt.subplots(figsize=(12, len(key_features) * 4), nrows=len(key_features), ncols=1, constrained_layout=True)
for i, feature in enumerate(key_features):
    try:
        PartialDependenceDisplay.from_estimator(model, X_test, [feature], ax=ax[i], grid_resolution=50)
        ax[i].set_title(f'Partial Dependence Plot for {feature}')
        ax[i].set_xlabel(feature)
        ax[i].set_ylabel('Predicted Probability of Self-Monitoring Recommendation')
    except Exception as e:
        ax[i].text(0.5, 0.5, f'Error: {str(e)}', ha='center', va='center')
        ax[i].set_title(f'Partial Dependence Plot for {feature} (Failed)')
        ax[i].set_xlabel(feature)
        ax[i].set_ylabel('')

plt.show()
```

```python
# Интерпретация
print("\nИнтерпретация влияния признаков на предсказание (на основе PDP):")
for feature in key_features:
    try:
        pdp_data = PartialDependenceDisplay.from_estimator(model, X_test, [feature], grid_resolution=50)
        pdp_values = pdp_data.pd_results[0]['average'][0]
        pdp_grid = pdp_data.pd_results[0]['grid_values'][0]
        slope = pdp_values[-1] - pdp_values[0]
        if slope > 0:
            direction = "увеличение признака повышает вероятность рекомендации самоконтроля АД"
        elif slope < 0:
            direction = "увеличение признака снижает вероятность рекомендации самоконтроля АД"
        else:
            direction = "признак имеет слабое или немонотонное влияние на вероятность рекомендации"
        print(f"{feature}: {direction}")
    except Exception as e:
        print(f"{feature}: Ошибка при интерпретации - {str(e)}")
```
